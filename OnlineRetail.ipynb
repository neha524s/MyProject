```python
def myfun(num):
    num=str(num)
    if num in ('9','1','6'):
        return 'B'
    elif num in ('8','4','2','7','0'):
        return 'A'
    elif num=='3' or num=='5':
        return 'C'
    else:
        return num
myfun('A')
```

```python
df['StockCode_cattt']=df['StockCode_cat'].apply(myfun)
```

```python
df.drop(columns=['StockCode_cat'],axis=1)
df.sample(10)
```
